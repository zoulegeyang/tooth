```python
# opencV代码块
import cv2
import tensorflow as tf
import os
import numpy as np

dir_mstar = 'D:\\QQdowanload\\2019-06-21 00-00-00-2019-08-20 12-00-00yuanpian\data\\3'

filenames = os.listdir(dir_mstar)
filelist = [os.path.join(dir_mstar, file) for file in filenames]
images2=[]
labels=[]
i=0
for file in filelist:
    
    aa=cv2.imread(file,0) 
    print(file)
    
    bb=cv2.resize(aa, (40, 40))
#     cv2.imshow('bb',bb)
    print(bb)#显示缩放后的图片
       
#     cv2.waitKey(0) 
    
    images2.append(bb)
    i+=1
    if i%2==0:
        labels.append(1)
    else:
        labels.append(0)
cc=np.array(images2,dtype='float')


labels=np.array(labels)

labels=labels.reshape(-1,1)
```

```python

from __future__ import print_function
from sklearn.model_selection import train_test_split
cc=cc.reshape(-1,40,40,1)
#该处由于没有labels所以就没有进行处理
train_images, test_images, train_labels, test_labels = train_test_split(cc, labels, test_size = 0.3)


# train_images, test_images= train_test_split(cc,test_size = 0.3)
print(train_images.shape,train_labels.shape)


import keras

from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D


import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"


epochs = 100
```

```python
# y_train = keras.utils.to_categorical(y_train, num_classes)
# y_test = keras.utils.to_categorical(y_test, num_classes)

model = Sequential()
model.add(Conv2D(40, (3, 3), padding='same',
                 input_shape=train_images.shape[1:]))
model.add(Activation('relu'))
model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
model.summary()
```

```python
opt = keras.optimizers.rmsprop(lr=0.0001, decay=1e-6)

# Let's train the model using RMSprop
model.compile(loss='binary_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])

train_images = train_images.astype('float32')
test_images = test_images.astype('float32')
train_images /= 255
test_images /= 255
model.fit(train_images, train_labels,
              batch_size=1,
              epochs=epochs,
              validation_data=(test_images, test_labels),
              shuffle=True)
```
